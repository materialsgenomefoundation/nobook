## Elastic, magnetic, and electric contributions to free energy

Chapters and focus on the thermal and hydrostatical pressure
contributions to Gibbs energy, which are the two prime variables
affecting phase stability in typical experimental environments. However,
there are other internal and external contributions, which are
particularly important for crystalline phases. Two important internal
contributions are from magnetic and electric polarizations of materials
with the corresponding external contributions due to the magnetic and
electric fields. Furthermore, for non-hydrostatic pressing of solid
phases, the $PV$ term in the combined law is to be replaced by elastic
energy calculated from elastic stress and elastic strain. The
corresponding works done to a system are as follows \[3\]

Eq. ‑
$dW_{elastic} = - V\sum_{i,j,k,l}^{}{\sigma_{ij}{d\varepsilon}_{kl}}$

Eq. ‑ $dW_{magnetic} = - V\sum_{i}^{}{H_{i}{dB}_{i}}$

Eq. ‑ $dW_{electric} = - V\sum_{i}^{}{E_{i}{dD}_{i}}$

where $i,j,\ k,\ l = 1,2,3$, $\sigma_{ij}$ and $\varepsilon_{kl}$ are
the components of stress and strain; $H_{i}$ and $B_{i}$ are the
components of magnetic field and magnetic induction; $E_{i}$ and $D_{i}$
are the components of electric field and electric displacement; and $V$
is the volume of the crystal. The negative sign in front of the equation
is due to the fact that the system does work to the surroundings when it
expands its volume due to the strain, magnetic induction, and electric
displacement.

Using the combined law of thermodynamics, can thus be re-written as
follows

Eq. ‑
$dU = TdS - V\left( \sum_{i,j,k,l}^{}{\sigma_{ij}{d\varepsilon}_{kl}} + \sum_{i}^{}{H_{i}{dB}_{i}} + \sum_{i}^{}{E_{i}{dD}_{i}} \right) + \sum_{}^{}\mu_{i}{dN}_{i} - Dd\xi$

The Legendre transformation similar to Helmholtz energy and Gibbs
energy, and , can be made to obtain the following characteristic free
energy functions

Eq. ‑ $dF = d(U - TS)$

$$= - SdT - V\left( \sum_{i,j,k,l}^{}{\sigma_{ij}{d\varepsilon}_{kl}} + \sum_{i}^{}{H_{i}{dB}_{i}} + \sum_{i}^{}{E_{i}{dD}_{i}} \right) + \sum_{}^{}\mu_{i}{dN}_{i} - Dd\xi$$

Eq. ‑ $dF_{H} = d\left( U - TS + \sum_{i}^{}{H_{i}B_{i}} \right)$

$$= - SdT - V\left( \sum_{i,j,k,l}^{}{\sigma_{ij}{d\varepsilon}_{kl}} - \sum_{i}^{}{B_{i}dH_{i}} + \sum_{i}^{}{E_{i}{dD}_{i}} \right) + \sum_{}^{}\mu_{i}{dN}_{i} - Dd\xi$$

Eq. ‑ $dF_{E} = d\left( U - TS + \sum_{i}^{}{E_{i}D_{i}} \right)$

$$= - SdT - V\left( \sum_{i,j,k,l}^{}{\sigma_{ij}{d\varepsilon}_{kl}} + \sum_{i}^{}{H_{i}{dB}_{i}} - \sum_{i}^{}{D_{i}{dE}_{i}} \right) + \sum_{}^{}\mu_{i}{dN}_{i} - Dd\xi$$

Eq. ‑
$dF_{EH} = d\left( U - TS + \sum_{i}^{}{H_{i}B_{i}} + \sum_{i}^{}{E_{i}D_{i}} \right)$

$$= - SdT - V\left( \sum_{i,j,k,l}^{}{\sigma_{ij}{d\varepsilon}_{kl}} - \sum_{i}^{}{B_{i}{dH}_{i}} - \sum_{i}^{}{D_{i}{dE}_{i}} \right) + \sum_{}^{}\mu_{i}{dN}_{i} - Dd\xi$$

Eq. ‑
$dG = d\left( U - TS + \sum_{i,j,k,l}^{}{\sigma_{ij}\varepsilon_{kl}} + \sum_{i}^{}{H_{i}B_{i}} + \sum_{i}^{}{E_{i}D_{i}} \right)$

$$= - SdT + V\left( \sum_{i,j,k,l}^{}{\varepsilon_{ij}d\sigma}_{kl} + \sum_{i}^{}{B_{i}dH_{i}} + \sum_{i}^{}{D_{i}{dE}_{i}} \right) + \sum_{}^{}\mu_{i}{dN}_{i} - Dd\xi$$

From above equations, it can be seen that the natural variables of
various free energies are
$F\left( T,\ \varepsilon_{ij},\ B_{i},\ D_{i},\ N_{i},\ \xi \right)$,
$F_{H}\left( T,\ \varepsilon_{ij},\ H_{i},\ D_{i},\ N_{i},\ \xi \right)$,
$F_{E}\left( T,\ \varepsilon_{ij},\ B_{i},\ E_{i},\ N_{i},\ \xi \right)$,
$F_{EH}\left( T,\ \varepsilon_{ij},\ H_{i},\ E_{i},\ N_{i},\ \xi \right)$,
and $G\left( T,\ \sigma_{ij},\ H_{i},\ E_{i},\ N_{i},\ \xi \right)$.
Clearly, there can be more combinations when the components of
$\varepsilon_{ij}$, $D_{i}$, and $B_{i}$ are partially replaced by their
conjugate potentials. The free energies listed above are useful
depending on how the system is constrained by the surroundings. For
practical applications, the elastic, magnetic, and electric properties
are usually considered for phases with fixed compositions, and at
equilibrium can then be written as

Eq. ‑
$dG = - SdT + V\left( \sum_{i,j,k.l}^{}{\varepsilon_{ij}d\sigma}_{kl} + \sum_{i}^{}{B_{i}dH_{i}} + \sum_{i}^{}{D_{i}{dE}_{i}} \right) + \mu dN$

The corresponding Gibbs-Duhem equation follows as below

Eq. ‑
$0 = - SdT + V\left( \sum_{i,j,k,l}^{}{\varepsilon_{ij}d\sigma}_{kl} + \sum_{i}^{}{B_{i}dH_{i}} + \sum_{i}^{}{D_{i}{dE}_{i}} \right) - Nd\mu$
